```python
from DataStructures.NeuralNetworks import RBF
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-1, 1, 50)
Y = np.sin((X + 0.5)**3)
Y_noised = np.sin((X + 0.5)**3) + np.random.randn(50) / 10
```

```python
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot()
ax.plot(X, Y)
ax.plot(X, Y_noised)
ax.title.set_text("y = sin(x+0.5)**3")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
rbf = RBF(K=6)
rbf.fit(X, Y)
```

```python
rbf.weights
```

```python
rbf_predict = rbf.predict(X)
```

```python
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot()
ax.plot(X, Y)
ax.plot(X, Y_noised)
ax.plot(X, rbf_predict)
ax.legend(["f(x)", "f(x) with noise", "RBF for f(x)"])
ax.title.set_text("f(x) = sin((x+0.5)**3)")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()
```
