```python
from DataStructures.NeuralNetworks import RBF
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-1, 1, 50)
Y = np.sin((X + 0.5)**3)
Y_noised = np.sin((X + 0.5)**3) + np.random.randn(50) / 10
```

## Plot function

```python
fig = plt.figure(figsize=(10,7))
ax = fig.add_subplot()
ax.plot(X, Y)
ax.plot(X, Y_noised)
ax.title.set_text("y = sin(x+0.5)**3")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

## RBF Network

```python
rbf = RBF(K=20)
rbf.fit(X, Y_noised)
```

```python
rbf.weights
```

```python
rbf_predict = rbf.predict(X)
```

```python
def plot_result(X, Y, Y_noised, rbf_predict, title="f(x) = sin((x+0.5)**3)"):
    fig = plt.figure(figsize=(10,7))
    ax = fig.add_subplot()
    ax.plot(X, Y, color="blue", linewidth=3)
    ax.plot(X, Y_noised, color="red", linewidth=3, linestyle='dotted')
    ax.plot(X, rbf_predict, color="black", linewidth=3, linestyle='dashed')
    ax.legend(["f(x)", "f(x) with noise", "RBF for f(x)"])
    ax.title.set_text(title)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    plt.show()
plot_result(X, Y, Y_noised, rbf_predict)
```

## Check RBF results for different number of joints

```python
def test_rbf(k=1):
    rbf2 = RBF(K=k)
    rbf2.fit(X, Y_noised)
    rbf2_predict = rbf2.predict(X)
    plot_result(X, Y, Y_noised, rbf2_predict, title=f"RBF for {k} joint(s)")
```

```python
for i in range(1, 10, 1):
    test_rbf(k=i)
```
